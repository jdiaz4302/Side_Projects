```python
sec_count = []

current_dict_index = 0
current_sec_count = 0

for i in lyrics_list:
    
    if (i == my_dict[current_dict_index]):
        sec_count.append(current_sec_count)
        current_sec_count = 0
        current_dict_index += 1
        print(i)
    else:
        current_sec_count += 1
        
    if (current_dict_index == len(my_dict)):
        break
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
pylab.rcParams['figure.figsize'] = (11, 8.5)
```

```python
import numpy as np

def calc_running_mean(x, N):
    """
        Unashamedly stolen from: 
        https://stackoverflow.com/questions/13728392/moving-average-or-running-mean
    """
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
running_window = 20
running_mean1 = np.concatenate((np.zeros(running_window - 1), calc_running_mean(sec_count, running_window)))
running_window = 10
running_mean2 = np.concatenate((np.zeros(running_window - 1), calc_running_mean(sec_count, running_window)))
running_window = 5
running_mean3 = np.concatenate((np.zeros(running_window - 1), calc_running_mean(sec_count, running_window)))
running_window = 2
running_mean4 = np.concatenate((np.zeros(running_window - 1), calc_running_mean(sec_count, running_window)))

plt.axvspan(0, 24, facecolor = 'red', alpha=0.20, label = 'Intro / Chorus')
plt.axvspan(80, 96, facecolor = 'red', alpha=0.20)
plt.axvspan(159, 174, facecolor = 'red', alpha=0.20)
plt.plot(running_mean1, c = 'black', alpha = 1.0, label = '20-word average')
plt.plot(running_mean2, c = 'grey', alpha = 0.5, label = '10-word average')
plt.plot(running_mean3, c = 'grey', alpha = 0.25, label = '5-word average')
plt.plot(running_mean4, c = 'grey', alpha = 0.125, label = '2-word average')
ax = plt.subplot(111)
ax.legend(bbox_to_anchor=(1.35, 0.67), fontsize = 12)
plt.title("Words per Second for Eminem's Rap God", fontsize = 16)
plt.xticks([0, 15, 30, 45, 60,
            75, 90, 105, 120,
            135, 150, 165, 180, 
            195, 210, 225, 240,
            255, 270, 285, 300,
            315, 330, 345, 360],
           ['0:00', '', '', '', '1:00',
            '', '', '', '2:00',
            '', '', '', '3:00',
            '', '', '', '4:00', 
            '', '', '', '5:00', 
            '', '', '', '6:00'],
          fontsize = 9)
plt.xlabel('Song Time', fontsize = 14);
```

```python
fastest_average = np.argmax(running_mean1)
fastest_instance = np.argmax(sec_count)

print(fastest_average)
print(fastest_instance)
```
