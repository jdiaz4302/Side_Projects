```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
def stupify_text(string):
    string_fixed = str()
    white_space_compensator = 0
    for i in range(len(string)):
        if (string[i] == " "):
            white_space_compensator += 1
        if ((i + white_space_compensator) % 2) != 0:
            string_fixed = string_fixed + string[i]
        else:
            string_fixed = string_fixed + string[i].upper()
    return(string_fixed)
```

```python
def plot_meme(file_path, text):
    img = mpimg.imread(file_path)
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.text(x = 0, y = -50, s = text,
             color = 'black', fontsize = 16,
             fontweight = 'bold')
```

```python
original_text = 'Do a real project, Jeremy!'

meme_text = stupify_text(original_text)

plot_meme('spongebob.jpg', meme_text)
```
